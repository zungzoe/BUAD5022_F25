# Capital Budgeting

```python
# import packages
import gurobipy as gp
from gurobipy import GRB
```

```python
# Set up the data
projects = ['A', 'B', 'C', 'D', 'E']

npv = {'A': 2,
       'B': 3,
       'C': 1,
       'D': 5,
       'E': 4}

budget = {'year1': 12,
          'year2': 8,
          'year3': 8,
          'year4': 4}

investments = {'year1': {'A': 3,
                         'C': 1,
                         'D': 10,
                         'E': 2},
               'year2': {'B': 5,
                         'C': 2,
                         'D': 4},
               'year3': {'B': 1,
                         'C': 1, 
                         'D': 2,
                         'E': 5},
               'year4': {'C': 2,
                         'E': 1} }
```

```python
# Create the model 
cap = gp.Model('capital_budgeting')
cap.ModelSense = GRB.MAXIMIZE
```

```python
# Create variables
x = cap.addVars(projects, vtype=GRB.BINARY, name='x')
cap.update()
```

```python
# Set the objective
cap.setObjective(gp.quicksum(npv[project]*x[project] for project in projects))
cap.update()
```
